```python
import sys 

rounds = [
    {
    'Shadow': {'kills': 2, 'assists': 1, 'deaths': True},
    'Blaze': {'kills': 1, 'assists': 0, 'deaths': False},
    'Viper': {'kills': 1, 'assists': 2, 'deaths': True},
    'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
    'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
    },
    {
    'Shadow': {'kills': 0, 'assists': 2, 'deaths': False},
    'Blaze': {'kills': 2, 'assists': 0, 'deaths': True},
    'Viper': {'kills': 1, 'assists': 1, 'deaths': False},
    'Frost': {'kills': 2, 'assists': 1, 'deaths': True},
    'Reaper': {'kills': 0, 'assists': 1, 'deaths': False}
    },
    {
    'Shadow': {'kills': 1, 'assists': 0, 'deaths': False},
    'Blaze': {'kills': 2, 'assists': 2, 'deaths': True},
    'Viper': {'kills': 1, 'assists': 1, 'deaths': True},
    'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
    'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
    },
    {
    'Shadow': {'kills': 2, 'assists': 1, 'deaths': False},
    'Blaze': {'kills': 1, 'assists': 0, 'deaths': True},
    'Viper': {'kills': 0, 'assists': 2, 'deaths': False},
    'Frost': {'kills': 1, 'assists': 1, 'deaths': True},
    'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
    },
    {
    'Shadow': {'kills': 1, 'assists': 2, 'deaths': True},
    'Blaze': {'kills': 0, 'assists': 1, 'deaths': False},
    'Viper': {'kills': 2, 'assists': 0, 'deaths': True},
    'Frost': {'kills': 1, 'assists': 1, 'deaths': False},
    'Reaper': {'kills': 1, 'assists': 1, 'deaths': True}
    }
]


def calcular_puntuacion(estadisticas):
    """Calcula la puntuación de un jugador."""
    kills = estadisticas['kills']
    assists = estadisticas['assists']
    deaths = 1 if estadisticas['deaths'] else 0  # 1 si murió, 0 si no
    return (kills * 3) + assists - deaths

def actualizar_datos_jugador(jugadores, jugador, estadisticas, puntuacion):
    """Actualiza las estadísticas acumuladas de un jugador."""

    if jugador not in jugadores:
        jugadores[jugador] = {'kills': 0, 'assists': 0, 'deaths': 0, 'puntos': 0, 'mvps': 0}

    jugadores[jugador]['kills'] += estadisticas['kills']
    jugadores[jugador]['assists'] += estadisticas['assists']
    jugadores[jugador]['deaths'] += 1 if estadisticas['deaths'] else 0
    jugadores[jugador]['puntos'] += puntuacion





def generar_ranking(rondas):
    """Genera el ranking de jugadores."""
    jugadores = {}

    for i, ronda in enumerate(rondas):
        puntuaciones_ronda = {}

        for jugador, estadisticas in ronda.items():
            puntuacion = calcular_puntuacion(estadisticas)
            puntuaciones_ronda[jugador] = puntuacion


            actualizar_datos_jugador(jugadores, jugador, estadisticas, puntuacion)
            
        mvp = max(puntuaciones_ronda, key=puntuaciones_ronda.get)
        jugadores[mvp]['mvps'] += 1
        
        print(f"Ranking ronda {i + 1}:")
        print("Jugador | Kills | Asistencias | Muertes | MVPs | Puntos")
        print("-" * 40)
        ranking_ronda = sorted(jugadores.items(), key=lambda item: item[1]['puntos'], reverse=True) 
        for jugador, estadisticas in ranking_ronda:
            estadisticas_ronda = ronda[jugador]
            kills = estadisticas_ronda['kills']
            assists = estadisticas_ronda['assists']
            deaths = 1 if estadisticas_ronda['deaths'] else 0
            mvps = jugadores[jugador]['mvps']
            puntos = jugadores[jugador]['puntos']
            print(f"{jugador} {kills} {assists} {deaths} {mvps} {puntos}")
        print("-" * 40)

    # ranking_final = sorted(jugadores.items(), key=lambda item: item[1]['puntos'], reverse=True)

    # print("\nRanking final:")
    # print("Jugador | Kills | Asistencias | Muertes | MVPs | Puntos")
    # print("-" * 40)
    # for jugador, estadisticas in ranking_final:
    #     print(f"{jugador} {estadisticas['kills']} {estadisticas['assists']} {estadisticas['deaths']} {estadisticas['mvps']} {estadisticas['puntos']}")
    # print("-" * 40)



generar_ranking(rounds)
```
